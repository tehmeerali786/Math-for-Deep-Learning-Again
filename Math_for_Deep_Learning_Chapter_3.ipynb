```python
import numpy as np

N = 10000000
B = 100
x = np.arange(B)/B

a = np.random.random(N)
a1 = np.histogram(a, bins=B)[0]
a2 = a1/a1.sum()
b = np.random.normal(0, 1, size=N)
b1 = np.histogram(b, bins=B)[0]
b2 = b1/b1.sum()
c = np.random.gamma(5.0, size=N)
c1 = np.histogram(c, bins=B)[0]
c2 = c1/c1.sum()
d = np.random.beta(5, 2, size=N)
d1 = np.histogram(d, bins=B)[0]
d2 = d1/d1.sum()
```

```python
import matplotlib.pyplot as plt

plt.hist(a, bins=100)
plt.hist(b, bins=100)
plt.hist(c, bins=100)
plt.hist(d, bins=100)
plt.show()
```

Central Limit Theorem

```python
# Regardless of the distribution; if various samples taken from the distribution, means of that samples will have normal curve
M = 10000
m = np.zeros(M)
for i in range(M):
  t = np.random.beta(5, 2, size=M)
  m[i] = t.mean()
print("Means of the means = %0.7f" % m.mean()) 
```

```python
h, x = np.histogram(m, bins=8)
h = h/h.sum()
plt.bar(x[:-1]+0.5*(x[1]-x[0]), h, width=0.8*(x[1]-x[0]))
plt.xlabel("Mean")
plt.ylabel("Probability")
plt.tight_layout(pad=0, w_pad=0, h_pad=0)
# plt.savefig("central_limit.png", dpi=300)
plt.show()
plt.close()
```

```python
# Law of Large Numbers => If you take from the sample from larger and larger sample, mean will close to population mean

m = []
for n in np.linspace(1, 8, 30):
  t = np.random.normal(1, 1, size=int(10**n))
  m.append(t.mean())

plt.plot(np.linspace(1, 8, 30), m)
plt.plot([1, 8], [1, 1], linestyle="--", color="k")
plt.plot("Exponent $10^n$")
plt.plot("Single sample mean")
plt.tight_layout(pad=0, w_pad=0, h_pad=0)
# plt.savefig("large_numbers.png", dpi=300)
plt.show()
```
